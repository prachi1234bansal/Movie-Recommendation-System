```
# Import libraries
import numpy as np
import pandas as pd
```

```
#Get the data
column_names=['user_id','item_id','rating','timestamp']
dataset=pd.read_csv('u.data',sep='\t',names=column_names)
dataset
```

```
column_names=['item_id','title']
movie_titles=pd.read_csv('u.item',sep='|',encoding="ISO-8859-1",names=column_names,usecols=range(2))
movie_titles.head()
```

```
dataset=pd.merge(dataset,movie_titles)
dataset
```

```
#EXPLORE THE DATA
#Lets explore the data a bit and get a look at some of the best rated movies.

```

```
#VISUALIZATION IMPORTS
```

```
import matplotlib.pyplot as plt
import seaborn as sb
sb.set_style('darkgrid')
%matplotlib inline
```

```
dataset.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```
dataset.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```
ratings=pd.DataFrame(dataset.groupby('title')['rating'].mean())
ratings
```

```
ratings['num of ratings']=pd.DataFrame(dataset.groupby('title')['rating'].count())
ratings
```

```
#Recommending Similar movies
```

```
moviemat=dataset.pivot_table(index='user_id',columns='title',values='rating')
moviemat
```

```
#Most rated movies
```

```
ratings.sort_values('num of ratings',ascending=False)
```

```

starwars_user_ratings=moviemat[input("enter")]
starwars_user_ratings
```

```
similar=moviemat.corrwith(starwars_user_ratings)
similar
```

```
corr_movie=pd.DataFrame(similar,columns=['Correlation'])
corr_movie.dropna(inplace=True)
corr_movie
```

```
corr_movie.sort_values('Correlation',ascending=False)
```

```
corr_movie[corr_movie['num of ratings']>100].sort_values('Correlation',ascending=False).head(5)
```
